## CAPA ADICIONAL DE VERIFICACI칍N

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.inspection import permutation_importance
import matplotlib.pyplot as plt

RANDOM_STATE = 42

# 0. Cargar la base
RUTA = r"C:\ITAM\MAESTRIA ITAM\Aprendizaje de M치quina\Examen Final\MutualFunds_limpio_variables_reducidas.csv"
df = pd.read_csv(RUTA)

print(df.shape)
print(df.columns)

# 1. Definir target y columnas
target = "morningstar_risk_rating"

# Categ칩ricas estrat칠gicas (nos aseguramos de que existan en el CSV)
cat_cols_candidatas = ["fund_category", "investment_type", "size_type", "region"]
cat_cols = [c for c in cat_cols_candidatas if c in df.columns]

# Todas las dem치s (excepto target y categ칩ricas) las tratamos como num칠ricas
numeric_cols = [c for c in df.columns if c not in cat_cols + [target]]

print("Columnas categ칩ricas usadas:", cat_cols)
print("N칰mero de variables num칠ricas:", len(numeric_cols))

X = df[numeric_cols + cat_cols]
y = df[target]
```

```python
# 1A. RandomForest para obtener ranking de importancia en num칠ricas
rf_importance = RandomForestClassifier(
    n_estimators=300,
    random_state=RANDOM_STATE,
    n_jobs=-1
)

rf_importance.fit(df[numeric_cols], y)

importancias = rf_importance.feature_importances_
feat_importances = (
    pd.DataFrame({
        "feature": numeric_cols,
        "importance": importancias
    })
    .sort_values("importance", ascending=False)
    .reset_index(drop=True)
)

print(feat_importances.head(20))
```

```python
# 1B. Evaluar F1 macro para distintos k (top-k variables num칠ricas)
from sklearn.metrics import f1_score, make_scorer

scorer = make_scorer(f1_score, average="macro")

resultados = []
k_values = list(range(10, min(len(numeric_cols), 80) + 1, 5))  # por ejemplo hasta 80 o todas si son menos

# Definimos la validaci칩n cruzada
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)

for k in k_values:
    top_k_numeric = feat_importances["feature"].iloc[:k].tolist()
    features_uso = top_k_numeric + cat_cols
    
    # Preprocesamiento por tipo de variable
    ct = ColumnTransformer(
        transformers=[
            ("num", SimpleImputer(strategy="median"), [c for c in features_uso if c in numeric_cols]),
            ("cat", Pipeline(steps=[
                ("imputer", SimpleImputer(strategy="most_frequent")),
                ("onehot", OneHotEncoder(handle_unknown="ignore"))
            ]), [c for c in features_uso if c in cat_cols])
        ]
    )
    
    # Modelo base (RF)
    model = RandomForestClassifier(
        n_estimators=300,
        random_state=RANDOM_STATE,
        n_jobs=-1
    )
    
    pipe = Pipeline(steps=[
        ("preprocess", ct),
        ("clf", model)
    ])
    
    scores = cross_val_score(pipe, df[features_uso], y, cv=cv, scoring=scorer, n_jobs=-1)
    
    resultados.append({
        "k_numeric": k,
        "num_features_total": len(features_uso),
        "f1_macro_mean": scores.mean(),
        "f1_macro_std": scores.std()
    })

res_k = pd.DataFrame(resultados)
print(res_k)

# 1C. Gr치fica del "codo"
plt.figure(figsize=(8,5))
plt.plot(res_k["k_numeric"], res_k["f1_macro_mean"], marker="o")
plt.xlabel("N칰mero de variables num칠ricas (top-k)")
plt.ylabel("F1 macro (validaci칩n cruzada)")
plt.title("Curva F1 macro vs n칰mero de variables num칠ricas (RF)")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    accuracy_score,
    f1_score,
    classification_report,
    confusion_matrix
)

import matplotlib.pyplot as plt
import seaborn as sns
import joblib

RANDOM_STATE = 42

# ==============================
# 1. Cargar base final
# ==============================
RUTA = r"C:\ITAM\MAESTRIA ITAM\Aprendizaje de M치quina\Examen Final\MutualFunds_base_final.csv"
df = pd.read_csv(RUTA)

print("Shape:", df.shape)
print(df.dtypes)
```

```python
# ==============================
# 2. Definir target y features
# ==============================
target = "morningstar_risk_rating"

# Nos aseguramos de que el target sea entero (clases 1...5)
df[target] = df[target].astype(int)

numeric_features = [
    "fund_return_category_rank_1year",
    "fund_beta_3years",
    "fund_stdev_3years",
    "fund_return_category_rank_ytd",
    "fund_beta_5years",
    "fund_stdev_5years",
    "fund_stdev_10years",
    "fund_r_squared_3years",
    "fund_return_category_rank_3months",
    "fund_return_2020_q1"
]

categorical_features = ["fund_category", "investment_type", "size_type"]

X = df[numeric_features + categorical_features].copy()
y = df[target].copy()

print("\nClases en y:")
print(y.value_counts().sort_index())
```

```python
# ==============================
# 3. Train / Test split
# ==============================
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=RANDOM_STATE
)

print("Train shape:", X_train.shape)
print("Test  shape:", X_test.shape)
```

```python
from sklearn.metrics import make_scorer

# ==============================
# 4. Preprocesamiento + modelo
# ==============================
numeric_transformer = SimpleImputer(strategy="median")

categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

preprocess = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_features),
        ("cat", categorical_transformer, categorical_features)
    ]
)

rf = RandomForestClassifier(
    random_state=RANDOM_STATE,
    n_jobs=-1
)

pipe_rf = Pipeline(steps=[
    ("preprocess", preprocess),
    ("clf", rf)
])

scorer = make_scorer(f1_score, average="macro")
```

```python
# ==============================
# 5. Grid de hiperpar치metros
# ==============================
param_grid = {
    "clf__n_estimators": [200, 300],
    "clf__max_depth": [None, 10, 20],
    "clf__min_samples_split": [2, 5, 10],
    "clf__min_samples_leaf": [1, 2, 4],
    "clf__max_features": ["sqrt", "log2"]
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)

grid_search = GridSearchCV(
    estimator=pipe_rf,
    param_grid=param_grid,
    cv=cv,
    scoring=scorer,
    n_jobs=-1,
    verbose=2
)

grid_search.fit(X_train, y_train)

print("\nMejores hiperpar치metros:")
print(grid_search.best_params_)
print("\nMejor F1 macro CV:", grid_search.best_score_)
```

# 4- MLP (RED NEURONAL MULTICAPA- SKLEARN)

```python
# ============================================================
# MODELO 4 CORREGIDO: MLP (Red Neuronal Multicapa) con escalado
# ============================================================

import pandas as pd

from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import (
    f1_score,
    accuracy_score,
    classification_report,
    confusion_matrix
)
from sklearn.metrics import make_scorer

import matplotlib.pyplot as plt
import seaborn as sns

RANDOM_STATE = 42

# ------------------------------------------------------------
# 1. Cargar base original
# ------------------------------------------------------------
RUTA = r"C:\ITAM\MAESTRIA ITAM\Aprendizaje de M치quina\Examen Final\MutualFunds_base_final.csv"
df = pd.read_csv(RUTA)

target = "morningstar_risk_rating"
df[target] = df[target].astype(int)

numeric_features = [
    "fund_return_category_rank_1year",
    "fund_beta_3years",
    "fund_stdev_3years",
    "fund_return_category_rank_ytd",
    "fund_beta_5years",
    "fund_stdev_5years",
    "fund_stdev_10years",
    "fund_r_squared_3years",
    "fund_return_category_rank_3months",
    "fund_return_2020_q1"
]

categorical_features = ["fund_category", "investment_type", "size_type"]

X = df[numeric_features + categorical_features].copy()
y = df[target].copy()

print("Shape X:", X.shape)
print("Distribuci칩n de clases en y:")
print(y.value_counts().sort_index())

# ------------------------------------------------------------
# 2. Train / Test split
# ------------------------------------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=RANDOM_STATE
)

print("\nTrain shape:", X_train.shape)
print("Test  shape:", X_test.shape)

# ------------------------------------------------------------
# 3. Preprocesamiento + MLP en Pipeline
#    - Imputaci칩n num칠rica
#    - Escalado num칠rico (StandardScaler)
#    - Imputaci칩n + OneHot en categ칩ricas
# ------------------------------------------------------------
numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

preprocess = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_features),
        ("cat", categorical_transformer, categorical_features)
    ]
)

mlp = MLPClassifier(
    max_iter=500,          # un poco m치s alto para evitar problemas de convergencia
    random_state=RANDOM_STATE
)

pipe_mlp = Pipeline(steps=[
    ("preprocess", preprocess),
    ("clf", mlp)
])

scorer = make_scorer(f1_score, average="macro")

# ------------------------------------------------------------
# 4. GridSearchCV para MLP
# ------------------------------------------------------------
param_grid_mlp = {
    "clf__hidden_layer_sizes": [(64,), (128,), (64, 32)],
    "clf__activation": ["relu", "tanh"],
    "clf__solver": ["adam"],
    "clf__alpha": [0.0001, 0.001],
    "clf__learning_rate_init": [0.001, 0.01]
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)

grid_mlp = GridSearchCV(
    estimator=pipe_mlp,
    param_grid=param_grid_mlp,
    scoring=scorer,
    cv=cv,
    n_jobs=-1,
    verbose=2
)

grid_mlp.fit(X_train, y_train)

print("\nMejores hiperpar치metros MLP:")
print(grid_mlp.best_params_)
print("Mejor F1 macro CV:", grid_mlp.best_score_)

# ------------------------------------------------------------
# 5. Evaluaci칩n en TEST
# ------------------------------------------------------------
best_mlp = grid_mlp.best_estimator_
y_pred_mlp = best_mlp.predict(X_test)

f1_mlp = f1_score(y_test, y_pred_mlp, average="macro")
acc_mlp = accuracy_score(y_test, y_pred_mlp)

print("\nResultados MLP en TEST:")
print("Accuracy:", acc_mlp)
print("F1 macro:", f1_mlp)
print("\nClassification report:")
print(classification_report(y_test, y_pred_mlp))

cm = confusion_matrix(y_test, y_pred_mlp)

plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Greens",
            xticklabels=sorted(y.unique()),
            yticklabels=sorted(y.unique()))
plt.xlabel("Predicho")
plt.ylabel("Real")
plt.title("Matriz de confusi칩n  MLP (con escalado en pipeline)")
plt.tight_layout()
plt.show()
```
